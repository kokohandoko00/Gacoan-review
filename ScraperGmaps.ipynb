```python
def get_review_summary(result_set):
    rev_dict = {'Review_Rate': [],
        'Review_Time': [],
        'Review_Text' : []}
    for result in result_set:
        review_rate = result.find('span', class_='ODSEW-ShBeI-H1e3jb')["aria-label"]
        review_time = result.find('span',class_='ODSEW-ShBeI-RgZmSc-date').text
        review_text = result.find('span',class_='ODSEW-ShBeI-text').text
        rev_dict['Review_Rate'].append(review_rate)
        rev_dict['Review_Time'].append(review_time)
        rev_dict['Review_Text'].append(review_text)
    import pandas as pd    
    return(pd.DataFrame(rev_dict))
```

# Word Embedding

## A. Word2Vec

```python
rev_list=[word_tokenize(Review_Text.lower()) for Review_Text in tqdm(df.Review_Text)]
```

```python
print(rev_list[:2])
```

## 1. CBOW Model

```python
model= Word2Vec(rev_list, vector_size=128,window=5,min_count=3, workers=4, sg=0,hs=0,epochs=1000)
```

### 1.1 Save

```python
model.save('E:/tugas koko/SEMESTER5/Natural Language Processing/Materi/Self Learning/model.w2v')
```

### 1.2 Load

```python
model=Word2Vec.load('E:/tugas koko/SEMESTER5/Natural Language Processing/Materi/Self Learning/model.w2v')
```

```python
from umap import UMAP
import numpy as np
import plotly.express as px
```

## B. Fast Text (Another method to transform word into vector)

Fast Text memiliki kemampuan untuk mengidentifikasi kata-kata yang sebenarnya tidak visibel dalam Vocab. Contoh, hanya terdapat kata "Katak", maka Fast Text dapat identifikasi Kata "Kat", "Tak", "Ka", dan "ak". Ini terjadi karena Fast Text menganggap bahwa setiap kata itu terdiri dari character n-gram.

```python
from gensim.models import FastText
```

```python
model_FastText= FastText(rev_list, vector_size=128,window=5,min_count=3, workers=4, sg=0,hs=0,epochs=1000)
```

```python
model_FastText.save('E:/tugas koko/SEMESTER5/Natural Language Processing/Materi/Self Learning/model_FastText.fasttext')
```

```python
model_FastText=FastText.load('E:/tugas koko/SEMESTER5/Natural Language Processing/Materi/Self Learning/model_FastText.fasttext')
```

```python
model_FastText.train(contoh,total_examples=len(contoh),epochs=100)
```

```python
FT=model_FastText.wv
```

```python
FT["manteuppp"]
```

```python
FT.similar_by_word("manteupp",topn=5)
```

```python
T=UMAP().fit_transform(FT.vectors)
```

```python
data_Fast=pd.DataFrame(T, columns=["umaps1","umaps2"])
data_Fast["text"]=FT.index_to_key
data_Fast
```

```python
fig=px.scatter(data_Fast, x="umaps1", y="umaps2", text="text")
fig.update_traces(textposition="top center")
fig.update_layout(
    height=800,
    title_text="Visualisasi FastText"
    )
fig.show()
```
